# Práctica 2 URDF Robots Móviles

## Objetivo

El objetivo de esta práctica es que el alumno pueda simular la navegación autónoma SLAM con un robot móvil diferencial (2,0) en un entorno simulado.

### Metas 

- Que el alumno aprenda las particularidades de los archivos urdf que describen a los robots móviles.
- Que el alumnos pueda configurar e implementar las características físicas de las llantas del robots para un entorno de simulación.
- Que el alumno puedo implementar el plug de control del robot diferencial.
- Que el alumno pueda implementar sobre el robot simulado un conjunto de sensores que permitan conocer el estado del robot y de su entorno para la navegación autónoma.
- Que el alumno aprenda a controlar el robot mediante la publiación sobre l tópico *cmd_vel*.
