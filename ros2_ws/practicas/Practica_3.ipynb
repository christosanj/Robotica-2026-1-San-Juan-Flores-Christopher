# Práctica 3 Nodos y Tópicos

## Objetivo

El alumno entendera las bases para programación de robots utilizano los Nodos y los Tópicos utilizando la librería de "ros2 client library python" y la librería "ros2 client library cpp", generando un Nodo *publicador* y un Nodo *subscriptor*


## Metas

- El alumno generará un espacio de trabajo de ROS 2 para administrar un paquete en Python y un paquete en C++.
- El alumno generará un Nodo en Python y un Nodo en CPP.
- El alumno programará un Nodo Publicador y el Tópico que se genera será leido por un Nodo Subscriptor en Python.
- El alumno programará un Nodo Publicador y el Tópico que se genera será leido por un Nodo Subscriptor en C++.

## Desarrollo de la práctica 

Dentro del espacio de trabajo de la clase, en la carpeta *src* se debe desarrollar los sigueintes elementos:


```text
 src/
    ├── nodes_topics_py/
    |       ├── nodes_topics_py/
    |       |       └── __init__.py   
    |       ├── scripts/
    |       │       ├── my_node.py
    |       │       ├── counter_node_py.py
    |       │       ├── number_pub_node_py.py
    |       │       ├── publihser_node.py
    |       │       ├── robot_pub_node.py
    |       │       └── Template_pythton_node.py
    |       ├── setup.cfg
    |       ├── setup.py
    |       └── package.xml
    └── nodes_topics_cpp/
            ├── src/
            |       ├── counter_node_cpp.cpp
            |       ├── my_node.cpp
            |       ├── number_pub_node_cpp.cpp
            |       ├── template_cpp.cpp
            |       ├── omnicontroller.launch.py
            │       └── counter_node_cpp.rviz
            ├── CMakeLists.txt
            └── package.xml
   
```

1. Se programará el paquete y los archivos en Python.
2. Se programará el paquete y los archivos en C++.
3. Se ejecutaran de manera simultanea los Publicadores y Subcriptores de ambas carpetas.
4. Con base en los programas anteriores el alumno generará un Nodo Publicador el cual publique la posición variante de un motor en radianes y un Nodo Publicador-Subcriptor que lea la posición enviada y la publique en radianes.
