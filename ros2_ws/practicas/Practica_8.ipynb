# Práctica 8 Implementación de sensores en un entorno simulación

## Objetivo

El objetivo de esta práctica es que el alumno pueda implementar sensores en un robot para conocer la información del estado del robot y el estado del entorno cuando este realiza una tarea determinada. 


### Metas 

- Que el alumno pueda implementar diferentes tipos de sensores en la descripción de un robot.
- Cofigurar los parámetros de los sensores para que sean concidentes con un sensor real.
- Tomar la lectura de los sensores mediante RVIZ2 y *rqt*.
