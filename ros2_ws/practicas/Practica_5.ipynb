# Práctica 5 Uso de Interface de Mensajes y Servicios

## Objetivo

El objetivo pueda generar interfaces personalizadas para el manejo básico e intermedio de messages y services.

### Metas 

- Que el alumno conozca los tipos de interface que maneja ROS 2 y como conocer sus características.
- Que el alumno sea capaz de generar y configurar interfaces personalizadas del tipo messages y services.
- Generar una interface message y utilizarla en un Nodo que publique la información del mensaje.
- Generar una interface service generar un service_node y un client_node que utilice el service creado.

# Previo de la Práctica 5


Se le solicita al alumno investigar y colocar la sigueinte información en esta sección:

- Investigatr los tipos de *interfaces* que maneja ROS 2.
- Investigar la manera en que funciona cada una de las *interfaces* de ROS 2. 
- Como se implementa la interface de message y la interface de services en Nodos programados en C++.


## Desarrollo de la práctica 

Para el desarrollo de la práctica se deben realizar las siguientes activiadades:

1. Generar un paquete de interface y paquete en el cual se utilizaran las interfaces.
2. Programar los Nodos en C++ que utilicen las interfaces personalizadas
3. Verificar el correcto funcionamiento de los Nodos.
