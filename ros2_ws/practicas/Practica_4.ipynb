# Práctica 4 Nodos y Topicos

## Objetivo

El objetivo de esta práctica es que el alumno aprenda a programar Nodos y Tópicos para aplicarlos en el control de robots y el manejo de informtación de todo tipo.

### Metas 

- Que el alumno aprenda a los aspectos básicos de la programación de Nodos de ROS 2 en Python y en C++.
- Que el alumnos maneje Nodos Publicadores y Nodos Subscriptores en Python, en un paquete en Python.
- Que el alumnos maneje Nodos Publicadores y Nodos Subscriptores en C++, en un paquete en cmake.
- Que el alumno aprenda a conectar la información entre Nodos.

### Contribución al perfil del egresado

La siguiente práctica contribuye en los siguientes puntos al perfil del egresado:

#### Aptitudes y habilidades

# Previo de la Práctica 4

Se le solicita al alumno investigar y colocar la sigueinte información en esta sección:

- Investigar la implementación de los Nodos en ROS 2.
- Inventigar cual es la estructura básica de un Nodo en ROS 2 programado en Python y programado en C++.
- Investigar cual es la estructura básica de paquete en Python.
- Investigar cual es la estructura básica de paquete en cmake.


## Desarrollo de la práctica 

Para el desarrollo de la práctica se deben realizar las siguientes activiadades:

1. Generar un paquete con políticas de Python.
2. Programar un Nodo básico, un Nodo Publicador y un Nodo Subcriptor en Python dentro del paquete en Python.
3. Configurar el paquete de Python.
4. Realizar las pruebas de funcionamiento de los Nodos programados en el paquete de Python.
5. Generar un paquete con políticas de cmake.
6. Programar un Nodo básico, un Nodo Publicador y un Nodo Subcriptor en C++ dentro del paquete en cmake.
7. Configurar el paquete de cmake.
8.Realizar las pruebas de funcionamiento de los Nodos programados en el paquete de cmake.
