# Práctica 7 Implementación de control de posición y de velocidades en juntas rotacionales

## Objetivo

El objetivo de esta práctica es que el alumno pueda configurar controles de posición y de velocidad en las juntas de un robot serial dentro de un entorno de simulación.

### Metas 

- Que el alumno aprenda aprenda configurar e implementar la libreria de control de ROS 2.
- Que el alumno sea capaz de configurar los controladores de las juntas de un robot dentro de la descripción del robot e implementarlas den Gazebo.
- Implementar la configuración de los controladores para el uso del gripper del robot.
- Implementar los controladores al iniciar una simulación dentro de un archivo *launch*

# Previo de la Práctica 7

Se le solicita al alumno investigar y colocar la sigueinte información en esta sección:

- Investigar la libreria de ROS Control.
- Investigar la configuración de control de posición y el control velocidad en ROS 2. 


## Desarrollo de la práctica 

Para el desarrollo de la práctica se deben realizar las siguientes activiadades:

1. Se configurarán los controladores de las juntas del robots.
2. Configurar las juntas especiales para las juntas para un gripper.
3. Configurar los controladores para que sean construidos por el control_manager de ROS.
4. Implementar los controladores para que sean generados en la simulación de Gazebo mediente el archivo launch.
5. Verificar la configuración del funcionamiento de las juntas.
